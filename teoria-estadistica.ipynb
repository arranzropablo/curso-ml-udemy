# Estadistica

$$ X = \{x_1,x_2,...,x_n\} $$<br>
$$ \lvert X\rvert = n $$<br>

Función de distribución: $$P(X \leq x)\ \ \text{i.e.: La probabilidad acumulada}$$

## Medidas de centralización:

Media aritmética: $$ \overline X = \frac{\sum_{i=1}^{n}x_{i}}{n}$$<br>

Mediana (Valor que cae en el medio): $$ P(X \leq m) = 0.5$$<br>

Moda (Valor más repetido): $$ P(X = M) \geq P(x = x_i) $$ <br> $$\forall i \ 1 \leq i \leq n$$<br>

Percentiles: $$ P(X \leq x_p) = p$$<br>$$ p \in [0,1]$$<br>

## Medidas de dispersión

Varianza: $$ s^2 = \frac{\sum_{i=1}^{n}(x_i - \overline x)^2 }{n - 1}$$<br>

Desviación típica: $$ s = +\sqrt{s^2}$$<br>

Coeficiente de variación: $$ CV = \frac{s}{\overline x} \cdot 100$$<br>

## Momento de orden r respecto de la media

El momento de orden 2 es casi igual que la varianza

$$ m_r = \frac{\sum_{i=1}^{n}(x_i - \overline x)^r }{n}$$



## Medidas de asimetría

Coeficiente de asimetría de Fisher: $$ CAF = \frac{\sum_{i=1}^{n}(x_i - \overline x)^3 }{n \cdot s^3}$$<br>
* Si sale 0 significa que es simétrica (hay la misma forma a la izquierda que a la derecha de la media)
* Si sale positivo la moda está a la izquierda de la media
* Si sale negativos la moda está a la derecha de la media<br><br>

Coeficiente de curtosis: $$ c = \frac{\sum_{i=1}^{n}(x_i - \overline x)^4 }{n \cdot s^4} - 3$$<br>
* Si sale 0 significa que es mesocúrtica (bien balanceado el cuerpo con las colas)
* Si sale positivo significa que es leptocúrtica (se concentran muchos valores en el medio y las colas son mas degradadas)
* Si sale negativo significa que es platicúrtica (pocos valores en torno a la media y muchos en las colas)<br><br>

## Muestreo aleatorio

Con estos cálculos inferimos los resultados para una población entera.

La media de las medias será un estimador de la media de la población.<br>
La desviación típica será la que tuviera la población dividida por $$ \sqrt{n} $$ Donde n es el tamaño de cada una de las muestras.

## Contraste de hipótesis

$$ 
    H_0: \text{Hipotesis nula} \\
    H_1: \text{Hipotesis alternativa}
$$

Contraste bilateral:

$$ \Bigg \{ \begin{aligned} 
             H_0: \mu = \mu_0 \\ 
             H_1: \mu \neq \mu_0
         \end{aligned}$$ 
         
Contrastes unilaterales:

$$ \Bigg \{ \begin{aligned} 
             H_0: \mu \leq  \mu_0 \\ 
             H_1: \mu \gt \mu_0
         \end{aligned}$$ 
         
$$ \Bigg \{ \begin{aligned} 
             H_0: \mu \geq \mu_0 \\ 
             H_1: \mu \lt \mu_0
         \end{aligned}$$
         <br><br>
La media que estamos incluyendo en nuestra hipótesis será:
$$ \mu_\overline X $$

#### Caso 1:
$$\sigma \text{ conocida}$$

Tenemos la desviación de la población conocida que será:

$$ \sigma_\overline X \rightarrow \frac{\sigma}{\sqrt{n}}$$

Aplicamos el TCL directamente (z-test):

$$ Z = \frac{\overline X - \mu_\overline X}{\sigma_\overline X} = \frac{\overline X - \mu}{\frac{\sigma}{\sqrt{n}}} $$

$$ \overline X \text{ es la media de una muestra, }\mu \text{ es la media de la hipotesis a validar, } \sigma \text{ es conocida y } n \text{ es el tamaño de la muestra}$$

#### Caso 2:
$$\sigma \text{ desconocida}$$

Primero estimamos la desviación típica:

$$ S = \frac{\sum{(X_i - \mu_\overline X)^2}}{n - 1} \rightarrow \sigma $$

Los datos se distribuyen según la distribución t de student:

$$ t = \frac{\overline X - \mu}{\frac{S}{\sqrt{n}}} $$

$$ \overline X \text{ es la media de una muestra, }\mu \text{ es la media de la hipotesis a validar, } S \text{ es lo estimado y } n \text{ es el tamaño de la muestra}$$

### Niveles de confianza

Son los umbrales o limites del valor de aceptación que estamos dispuestos a aceptar para decir que la hipotesis nula es verdadera.<br>
Si para la media de la hipotesis nula definimos un intervalo de confianza centrado en él y el valor del estimador cae dentro del rango aceptaremos la hipótesis nula.<br>

Por ejemplo, en una distribución con media: mu y desviación: sigma:
* Hay un nivel de confianza del 99.7% de que la variable aleatoria estará en el rango +-3 sigma
* Hay un nivel de confianza del 95.4% de que la variable aleatoria estará en el rango +-2 sigma
* Hay un nivel de confianza del 68.2% de que la variable aleatoria estará en el rango +-1 sigma
<br><br>

Hay que definir el nivel de significación que queremos (1-p --> alpha). Esto representa la probabilidad de que la hipótesis nula **NO** sea cierta.<br>
Suele ser del orden de 0.1, 0.01, 0.05...

Obtenido el Z0 (del z-test o del t-test), se define el p-valor como la probabilidad de que la función de distribución (F(X)) supere dicho valor. I.e.: es la probabilidad de caer por encima de Z0 (Si p-valor > alpha --> aceptamos hipótesis nula, si p-valor < alpha --> aceptamos hipótesis alternativa).

**Nota para aclararme:** Como alpha delimita la región de rechazo, si el z-valor cae fuera de esta región de rechazo (i.e.: p-valor es mayor que alpha) entonces aceptamos la hipótesis nula.

### Tests de cola por la izquierda, de cola por la derecha o de dos colas

Como se ve en la sección de este notebook que se llama contraste de hipótesis hay 2 tipos de contrastes:
* Bilateral                          (Test de 2 colas)
* Unilateral
    * Hipótesis alternativa mayor    (Test de cola por la derecha)
    * Hipótesis alternativa menor    (Test de cola por la izquierda)
    
    <br><br>
    
#### Test de cola por la izquierda
Aceptaremos la hipótesis nula y rechazaremos la alternativa si el p-valor es mayor que el nivel de significación o si el z-valor de contraste es mayor al z-valor que corresponda al nivel de significación.<br>
En caso contrario rechazamos.<br>

#### Test de cola por la derecha
Aceptaremos la hipótesis nula y rechazaremos la alternativa si el p-valor es mayor que el nivel de significación o si el z-valor de contraste es menor al z-valor que corresponda al nivel de significación.<br>
En caso contrario rechazamos.<br>

#### Test de dos colas
El nivel de significación se reparte en alpha/2 en la cola izquierda y en alpha/2 en la cola derecha.<br>
Aceptamos o rechazamos con las mismas "reglas" que en los otros dos tests anteriores (comprobando los dos, si uno de los dos nos hacen rechazar la hipótesis nula, la rechazamos)<br><br>

**p-valor es un area(la probabilidad de que la función de distribución supere el valor de Z) por lo tanto en todos los casos tiene que ser mayor que el área alpha, que es la probabilidad de que la función de distribución caiga en la región crítica)**<br><br>



### Pasos de un contraste de hipótesis

1. Definir una hipótesis nula $\mu_0$ y una hipótesis alternativa<br>
2. Tomar una muestra aleatoria de tamaño n y calcular el valor del estimador.<br>
3. Calcular el estadístico de contraste (z-valor, t-valor)<br>
4. Calcular el p-valor asociado (la probabilidad de que la distribución sea mayor o menor que el z-valor)<br>
5. Comparar p-valor y el nivel de significación.

## Test de Chi cuadrado

#### Bondad de ajuste (1 variable)
$H_0$ será suponer que los datos siguen una determinada distribución y en base a esa hipótesis calculamos cuales serían los valores esperados.<br>
#### Test de homogeneidad (2 variables)

#### Test de independencia (2 variables)
$H_0$ será suponer que las 2 variables no están relacionadas.<br><br>

En todos estos casos el valor se obtiene con esta formula:<br>
$$\chi^2 = \sum_{i=1}^{n} \frac{(O_i - E_i)^2}{E_i}$$<br>
$O_i$ son los observados y $E_i$ son los esperados
<br><br>

Dependiendo de los grados de libertad (df = grados de libertad) la chi cuadrado tendrá una forma diferente.<br>
En la bondad de ajuste $df = n_{categorias} - 1$<br>
En el test de independencia $df = (n_{filas} - 1) \cdot (n_{columnas} - 1)$


### Ejemplo de prueba de bondad de ajuste

Tomamos una moneda que no sabemos si está o no está trucada:
$$ \Bigg \{ \begin{aligned} 
             H_0: \text{La proporción de caras y cruces es 0.5} \\ 
             H_1: \text{La proporción de caras y cruces no es 0.5}
         \end{aligned}$$ 
         
y $\alpha = 0.05$.<br><br>

Tras 1000 lanzamientos obtenemos 553 caras y 447 cruces.<br><br>

$$\chi^2 = \frac{(553 - 500)^2}{500} + \frac{(447 - 500)^2}{500} = 11.236$$<br>
Como hay 2 categorias --> df = 1<br><br>

Con estos datos consultamos la tabla de $\chi^2$ y vemos que: <br><br>

$$\chi^{2}_{\alpha} = \chi^{2}_{0.05} = 3.841 < \chi^{2} = 11.236$$<br><br>

Vemos que nuestro valor de $\chi^2$ entra dentro de la región crítica así que rechazamos la hipótesis nula.<br>
Concluimos que la moneda está trucada y la hipótesis alternativa es cierta.

## Correlación

Si existe una correlación entre dos variables existe cierta dependencia entre ellas.<br>
Por tanto, la correlación es una función matemática que relaciona dos variables.<br><br>
Ejemplo de correlación:
#### Correlación lineal
$$ y = ax + b$$

#### Correlación exponencial
$$ y = e^{ax} + b$$


### Coeficiente de correlación de Pearson

$$ r = \frac{\sum_{i=1}^{n}{(x_i - \overline{x})(y_i - \overline{y})}}{\sqrt{\sum^{n}_{i=1}{(x_i - \overline{x}) ^2}\sum^{n}_{i=1}{(y_i - \overline{y}) ^2}}}$$

Valores $\in [-1,1]$.<br>
Una correlación positiva significa que habrá una correlación directa y una negativa que habrá una correlación inversa.<br>
Si son valores cercanos al 0 (incluso 0.5, 0.6...) puede que sea mera casualidad.<br><br>

** La correlación no siempre implica la causalidad (puede que este relacionada pero influyan más factores.**

### Ejemplo de correlación


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("./datasets/ads/Advertising.csv")
data.head()
```

```python
def getCorrelation(col1, col2):
    numerador = sum((data[col1] - data[col1].mean()) * (data[col2] - data[col2].mean()))

    denominador = np.sqrt(sum((data[col1] - data[col1].mean())**2)*sum((data[col2] - data[col2].mean())**2))

    return numerador / denominador
```

```python
print("Correlación TV-Sales:", getCorrelation("TV", "Sales"))

print("Correlación Radio-Sales:", getCorrelation("Radio", "Sales"))

print("Correlación Newspaper-Sales:", getCorrelation("Newspaper", "Sales"))
```
