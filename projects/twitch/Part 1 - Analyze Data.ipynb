Create a new column named genre for each of the games.

Group the games into their genres: Multiplayer Online Battle Arena (MOBA), First Person Shooter (FPS), Survival, and Other.

```python
def get_genre(val):
    if (val in ['League of Legends', 'Dota 2', 'Heroes of the Strom']):
        return 'MOBA'
    elif val in ['Counter-Strike: Global Offensive']:
        return 'FPS'
    elif val in ['DayZ', 'Survival Evolved']:
        return 'Survival'
    else:
        return 'Other'
    
df_stream['genre'] = df_stream.game.apply(get_genre)

df_stream.head()

#Other way https://stackoverflow.com/questions/19913659/pandas-conditional-creation-of-a-series-dataframe-column
```

Let’s write a query that returns three columns:

The hours of the time column

The view count for each hour

Lastly, filter the result with only the users in your country.

```python
from datetime import datetime

df_stream['hour'] = df_stream.apply(lambda row: datetime.strptime(row.time, '%Y-%m-%d %H:%M:%S').hour, axis=1)
```

```python
df_stream[df_stream.country == 'ES'].groupby('hour').game.count()
```
