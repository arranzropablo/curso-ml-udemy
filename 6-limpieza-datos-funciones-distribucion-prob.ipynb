## Distribución normal
Según obtengamos datos se irá formando una distribución normal.<br>
Podemos obtener números aleatorios que sigan una distribución normal.

```python
data = np.random.randn(10000)
#data = np.random.randn(2,4) #dims

%matplotlib inline
plt.plot(range(1,10001), sorted(data))  #esto muestra la función de distribución acumulada
```

Si queremos una distribución normal que no sea la estandar (media = 0, desviación = 1) tenemos que hacer lo contrario a tipificar (Teorema Central del Limite, sirve para convertir cualquier variable en una normal estandar), que es restar la media y dividir entre la desviación (lo contrario es multiplicar y sumar):

```python
media = 5
sd = 3
data = media + sd * np.random.randn(10000)
#Z = (X - mu)/sd --> N(0,1)
#X = Z * sd + mu

%matplotlib inline
plt.plot(range(1,10001), sorted(data))
```

## Simulación de Monte Carlo
Calculamos π teniendo un circulo encerrado en un cuadrado de lado 2r siendo r el radio del circulo.<br>
La probabilidad de que un punto caiga en los límites del círculo es de π/4 (área del círculo / área del cuadrado)<br>

1. Generamos 2 números aleatorios entre 0 y 1 (este sería un punto al azar en el cuadrado)
2. Calcular sqrt(x \* x + y \* y) (i.e.: La distancia euclídea al centro).
3. Si ese cálculo es menor que 1 entonces estamos en el círculo.

Hacemos esto muchas veces y calculamos la probabilidad de caer dentro del círculo.<br>
Multiplicaremos este número por 4 y será el valor de π.

```python
np.random.seed(2018)

n_puntos = 10000
n_reps = 100

pi_value_list = []

for i in range(n_reps):
    x = np.random.uniform(0,1,n_puntos)
    y = np.random.uniform(0,1,n_puntos)
    n_inside_circle = 0
    
    for j in range(n_puntos):
        if(np.sqrt(x[j] * x[j] + y[j] * y[j]) <= 1):
            n_inside_circle += 1
            
    pi_value_list.append(4 * float(n_inside_circle) / n_puntos)
    
sum(pi_value_list) / n_reps
```
