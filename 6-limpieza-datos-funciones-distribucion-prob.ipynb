# Funciones de distribución de la probabilidad
## Distribución uniforme
En esta distribución todos los números tienen igual probabilidad de aparición como podemos ver en la gráfica.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.uniform(1,100, 1000000)
data

%matplotlib inline
plt.hist(data)
```

## Distribución normal
Según obtengamos datos se irá formando una distribución normal.<br>
Podemos obtener números aleatorios que sigan una distribución normal.

```python
data = np.random.randn(10000)
#data = np.random.randn(2,4) #dims

%matplotlib inline
plt.plot(range(1,10001), sorted(data))  #esto muestra la función de distribución acumulada
```

Si queremos una distribución normal que no sea la estandar (media = 0, desviación = 1) tenemos que hacer lo contrario a tipificar (Teorema Central del Limite, sirve para convertir cualquier variable en una normal estandar), que es restar la media y dividir entre la desviación (lo contrario es multiplicar y sumar):

```python
media = 5
sd = 3
data = media + sd * np.random.randn(10000)
#Z = (X - mu)/sd --> N(0,1)
#X = Z * sd + mu

%matplotlib inline
plt.plot(range(1,10001), sorted(data))
```
