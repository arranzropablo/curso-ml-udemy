# Agrupación de datos

## Agrupación de datos por categoría

Crearemos un dummy dataset con variables categóricas y numéricas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

n = 500

gender = ["Male", "Female"]
income = ["Poor", "Middle class", "Rich"]
gender_data = [np.random.choice(gender) for i in range(n)]
income_data = [np.random.choice(income) for i in range(n)]

height = []
weight = []
age = []
```

La altura y el peso suelen seguir una distribución normal.<br>
Podemos asumir que la altura tiene de media 160cm y una desviación de 30cm.<br>
Podemos asumir que el peso tiene de media 65kg y una desviación de 25kg.<br>
La altura seguirá una distribución normal (Aunque debería ser uniforme).<br><br>
También genera unos datos "income" con una distribución normal a los que no les veo mucho sentido porque deberían ir acordes a "income_data" (por ejemplo, no debería ganar más una persona "Poor" que una persona "Rich") así que no generaré esta columna.

```python
height = 160 + 30 * np.random.randn(n)
weight = 65 + 25 * np.random.randn(n)
age = np.random.uniform(1,100,n).astype(np.int64)  #el astype es para tener edades enteras no float

data = pd.DataFrame(
    {
        "Gender": gender_data,
        "Income": income_data,
        "Height": height,
        "Weight": weight,
        "Age": age
    }
)

data.head()
```

```python
gender_group = data.groupby("Gender")

for name, group in gender_group:
    print(name, "\n", group)
```

```python
gender_group.get_group("Female")    #Si queremos solo un grupo
```

También podemos agrupar por dos parámetros

```python
double_group = data.groupby(["Gender", "Income"])

for name, group in double_group:
    print(name, "\n", group)
```

```python
#Como se trata de un dataframe podemos hacer cualquier operación sobre él
double_group.describe()
double_group.sum()
double_group.size()
double_group.mean()
```

Para variables numéricas podemos pedirle algo específico sobre cada una (siguiendo agrupadas por categorías)

```python
double_group.aggregate(
    {
        "Weight": np.sum,
        "Age": np.mean,
        #Si queremos definir nuestra propia "función" de agregado:
        "Height": lambda h: np.mean(h)/np.std(h)
    }
)
```

Si queremos aplicar la misma función a todas las columnas:

```python
double_group.aggregate([np.mean, np.std, lambda x: np.mean(x)/np.std(x)])
```

## Filtrado de datos

```python
#Esto devuelve los elementos que pertenecen a un grupo cuya media de edad es mayor que 50
double_group["Age"].filter(lambda x: x.mean() > 50)
```

## Transformación de variables

Para, por ejemplo, tipificar los dato o sustituir los valores que falten en el dataset

```python
zscore = lambda x: (x - x.mean()) / x.std()
z_group = double_group.transform(zscore)

#Lo convierte en una normal (0,1)
plt.hist(z_group["Weight"])
```
